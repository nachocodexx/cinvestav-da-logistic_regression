## Función sigmoid o logit:

$$
    p(y=1|\hat{x})=\frac{1}{1+exp(-\hat{\mathbf{w}}^{T}\mathbf{\hat{x}})}
$$

```python
def sigmoid(x):
    return 1/(1+np.exp(-x))
```

## Descenso de gradiente

```python
def gradient_descent(X,y,**kwargs):
    LAMBDA       = kwargs.get("LAMBDA",1/(10**3))
    epochs       = kwargs.get("epochs",100)
    LR           = kwargs.get("learning_rate",0.001)
    threshold    = kwargs.get("threshold",1/(np.power(10,3)))
    show_entropy = kwargs.get("show_entropy",False)
    w          = np.random.rand(5).reshape(-1,1)
    predicate  =  True
    i          = 0 
    while(predicate):
#       Linear discriminant function
        x                   = (w.T@X)
#       Mapping values to (0,1)
        _y                  = sigmoid(x)
#       Supervised learning  
        yy                  = (_y-y)
        delta_Lw            = X@(yy.T)
#       w(i) vector of weights at iteration k
        w_temp              = w 
#       w(i+1) new vector of weights
        w                   = w-(LR*delta_Lw) + LAMBDA*w
#       cross - entropy 
        Lw                  = cross_entropy_loss(w,y,_y,l=LAMBDA)
        if(show_entropy):
            print("[{}]- CROSS ENTROPY LOSS: {}".format(i,Lw))
#       check exit conditions: Max iteration reached or Convergence
        norm_diff           = LA.norm(np.abs(w_temp-w))
        max_iter_reached    = epochs-1<=i
        convergence_reached = norm_diff < threshold
        predicate = not(( max_iter_reached or convergence_reached ))
#       Increment the counter  
        i+=1 
    return {'w':w,
            'max_iter_reached':max_iter_reached,
            'convergence_reached':convergence_reached,
            "iterations":i
           }
```

# Práctica #2: Regresión softmax

a) Implementar el algoritmo de regresión softmax usando la regla de actualización con descenso de gradiente

Softmax:

$$
\hat{y}_{ij} = \frac{\text{exp}(\mathbf{\hat{w}^T\hat{x}_i} )}{\sum_{l=1}^{c} \text{exp}(\mathbf{\hat{w}_{l}^{T} \hat{x}_{i} })}
$$

```python
# Softmax for a specific class
def _softmax(xi,ws,i):
    denominator = Vector(*list(map(lambda wi:np.exp(wi.T@xi),ws.T))).sum()
    return np.exp(ws[:,i].T@xi) /denominator
# Calculate softmax for c classes.
def __softmax(X,ws,**kwargs):
    c = kwargs.get("c",3)
    results = []
    for xi in X:
        results.append(Vector(*list(map(lambda i: _softmax(xi,ws,i),range(c)))))
    return Vector(*results)
```

### Función de entropía cruzada categórica

$$
\mathcal{L}(\mathbf{\hat{w_{1}},....,\hat{w_{c}}}) = - \sum_{i=1}^{n} \sum_{j=1}^{c} y_{ij}ln(\hat{y}_{ij})
$$

```python
def categorical_cross_entropy(X,Y,W):
    return -(Y*np.log(__softmax(X.T,W))).sum()
```

### One-hot enconding

```python
def toOneHotEncoding(y):
    vLen   = np.unique(y).shape[0]
    def __change(yi):
        tmp     = np.zeros(vLen)
        tmp[yi] = 1 
        return tmp.astype(int) 
    oneHot = Vector(*list(map(__change,y)))
    return oneHot
```

### Decenso de grandiente

```python
def gradient_descent_multi(X,y,**kwargs):
    epochs     = kwargs.get("epochs",100)
    LR         = kwargs.get("learning_rate",0.001)
    threshold  = kwargs.get("threshold",1/(np.power(10,3)))
    LAMBDA     = kwargs.get("l",0.0001)
#     W          = np.random.rand(d,c)
    W          = np.zeros([X.shape[0],3])
    Y1_train   = toOneHotEncoding(y)
    i = 0 
    predicate=True
    Lws = []
    while predicate:
        _Y       = __softmax(X.T,W)
        delta_Lw = X@(_Y-Y1_train)
        W_k      = W 
        W        = (W_k - LR*delta_Lw) + (LAMBDA*W_k)
        Lw       = categorical_cross_entropy(X,Y1_train,W)
        Lws.append(Lw)
        
        norm     = LA.norm(W_k-W,axis=0)
        convergence_reached = np.all(norm<threshold)
        max_iter_reached    =  epochs<=i
        predicate = not(( max_iter_reached or convergence_reached ))
        
        i+=1 
    return [W,convergence_reached,max_iter_reached,i]
```

## Decesnso de gradiente

```python
def cost_fn(X,w):
    return (-w.T@X).sum()
def GD(X,**kwargs):
    w             = np.zeros(X.shape[0]).reshape(-1,1)
    lr            = kwargs.get('learning_rate',0.001)
    epochs        = kwargs.get('epochs',100)
    show_progress = kwargs.get('show_progress',False )
    jpws          = []
    for i in range(epochs):
        gxs       = w.T@X
        mask0     = (gxs >0).T.flatten()
        mask1     = (gxs <=0).T.flatten()
        _X_       = X[:,mask1]
        Jpw       = cost_fn(_X_,w)
        jpws.append(Jpw)
        delta_Jpw = _X_.sum()
        w         = w + (lr*delta_Jpw)
        if(show_progress):
            print("[{}] Jpw = {}".format(i,Jpw))
    return w,jpws
_w, jpws = GD(XW,epochs=10000,learning_rate=0.00001)
_w
```
